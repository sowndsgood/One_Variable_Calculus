```python
#Importing Libraries
import numpy as np
```

```python
#True Value of area
actual=np.pi/4
print(actual)
```

## 1. Reimann Sum (Numerical Approximation)

```python
def f(x):  #Function of circle
  return (1-x**2)**(1/2)

n=1000     #Number of rectangles or sub-intervals
delta_x=1/n   #Width of rectangles
x_values=np.linspace(0,1,n)  #To generate values with same intervals with Linear Space function
reimann_sum=sum(f(x)*delta_x for x in x_values)  #Formula for Reimann Sum
print(reimann_sum)
```

```python
error1=actual-reimann_sum   #Error of area by Reimann Sum
print(error1)
```

## 2. Geometry (Sector of a Circle)

```python
r=1     #radius of circle
angle_sector=np.pi/2  #angle of sector is 90 degree
angle_circle=2*np.pi  #angle of circle is 360 degree
area_circle=np.pi*r*r  #area of circle is pi*r*r

area_sector=(angle_sector/angle_circle)*area_circle
print(area_sector)
```

```python
error2=actual-area_sector  #Error would be zero
print(error2)
```

## 3. Integral Calculus (Exact Area Calculation)

### Using Libraries:

#### 1. SymPy

```python
import sympy as sp #Importing SymPy library for symbolic mathematics

x=sp.Symbol('x')  #Defining 'x' as a symbolic variable
f=sp.sqrt(1-x**2)  #Define the function f(x) = sqrt(1 - x^2)
lower_limit=0  #Lower limit of integration
upper_limit=1  #Upper limit of integration
area_sympy=sp.integrate(f,(x,lower_limit,upper_limit))
print(area_sympy)
```

```python
error3=actual-area_sympy   #Error of area by SymPy
print(float(error3))
```

#### 2. SciPy

```python
from scipy import integrate

def f(x):   #Define the function f(x) = sqrt(1 - x^2)
    return np.sqrt(1-x**2)

lower_limit=0   #Lower limit of integration
upper_limit=1   #Upper limit of integration
area_scipy,error_estimate=integrate.quad(f,lower_limit,upper_limit)  #Integration using SciPy's quad function
print(area_scipy)
```

```python
print(error_estimate)  #Printing the error estimate of integration
```

```python
error4=actual-area_scipy   #Error of area by SciPy
print(error4)
```

Analysis:

```python
abs(error1)>abs(error3)   #Error by Reimann Sum is greater than Area by SymPy
```

```python
abs(error1)>abs(error4)  #Error by Reimann Sum is greater than Area by SciPy
```

```python
abs(error3)<abs(error4)  #Error by SciPy is greater than Area by SymPy
```
